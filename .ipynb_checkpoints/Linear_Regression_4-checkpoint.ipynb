```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datas/Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop('Sales' , axis=1)
X = X.drop('Unnamed: 0' , axis=1)
X
```

```python
y = df['Sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# we preprocess our data first
polynomial_converter = PolynomialFeatures(
    degree=3 , include_bias=False
)
```

```python
# now we have new features
polynomial_features = polynomial_converter.fit_transform(X)
```

```python
polynomial_features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# time to test and train 
X_train , X_test , y_train , y_test = train_test_split(
    polynomial_features ,
    y , 
    test_size=0.3,
    random_state=101
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# time to scale our data
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
X_train_s[0]
```

```python
X_train_s[: , 0].std()
```

```python
# untill now we have learned Regularaztion and Standarzation , time to Ridge Regrassion
```

```python
# Ridge Regrassion
# Ridge Regression is commonly used in scenarios where:
# Multicollinearity is present: When predictor variables are highly correlated.
# Overfitting is a concern: To reduce the risk of overfitting by adding a penalty
# term to the loss function More features than observations: When the number of predictor
# variables exceeds the number of observations Ridge vs. Lasso: Use Ridge when you want
# to include all features but reduce their impact, whereas Lasso is better for feature 
# selection as it can shrink some coefficients to zero
# *** Tips ***
# Standardize Features: It's crucial to standardize your features before applying Ridge
# Regression to ensure the penalty is applied uniformly Choose Lambda (λ):
# The value of λ (the penalty term) is critical. Use cross-validation to find the optimal
# value 
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train_s , y_train)
```

```python
test_predictions = ridge_model.predict(X_test_s)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
MAE = mean_absolute_error(y_test , test_predictions)
MSE = mean_squared_error(y_test , test_predictions)
RMSE = np.sqrt(MSE)
r2_score  =r2_score(y_test , test_predictions)
```

```python
ridge_cv_model = RidgeCV(alphas=(0.1, 1.0 , 10.0), scoring="neg_mean_absolute_error")
```

```python
ridge_cv_model.alpha_
```

```python
test_predictions = ridge_cv_model.predict(X_test_s)
```

```python
MAE = mean_absolute_error(y_test , test_predictions)
MSE = mean_squared_error(y_test , test_predictions)
RMSE = np.sqrt(MSE)
```

```python
# **** Lasso Regrassion ****
# 1.Use Cases
# Lasso Regression is commonly used when:
# Feature Selection: It's great for reducing the number of
# predictor variables by shrinking some coefficients to zero.
# High-Dimensional Data: When you have more predictors than observations.
# Model Simplification: To simplify models and make them more interpretable.
# Important Tips About Lasso Regression
# Standardize Features: Standardize your features before applying Lasso
# Regression to ensure the penalty is applied uniformly.
# Choose Lambda (λ): The value of λ (the penalty term) is critical.
# Use cross-validation to find the optimal value.
# Interpretation: Understand that Lasso can produce sparse models
# with some coefficients being zero, which simplifies interpretation but might
# exclude some relevant features.
```
