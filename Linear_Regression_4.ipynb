```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# feature Scaling - KNN
# Normalazation - Standardzation
```

```python
df = pd.read_csv("datas/Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop('Sales' , axis=1)
X = X.drop('Unnamed: 0' , axis=1)
X
```

```python
y = df['Sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_converter = PolynomialFeatures(degree=3 , include_bias=False)
```

```python
poly_features = poly_converter.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(
    poly_features , 
    y , 
    test_size=0.3,
    random_state=101
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_s = scaler.transform(X_train)
```

```python
X_test_s = scaler.transform(X_test)
```

```python
X_train_s[0]
```

```python
X_train_s[:, 0].std()
```

```python
# find best alpha (a;pha helps us againts overfitting and underfitting)
from sklearn.linear_model import RidgeCV
```

```python
ridge_cv_model = RidgeCV(alphas= (0.1 , 1.0 , 11.0) , scoring="neg_mean_absolute_error")
```

```python
ridge_cv_model.fit(X_train_s , y_train)
```

```python
ridge_cv_model.alpha_
```

```python
test_predictions = ridge_cv_model.predict(X_test_s)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, test_predictions)
MSE = mean_squared_error(y_test , test_predictions)
RMSE = np.sqrt(MSE)
```

```python
ridge_cv_model.coef_
```
